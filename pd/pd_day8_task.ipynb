```python
import pandas as pd
import matplotlib.pyplot as plt


# 1. 테슬라

tsla_df = pd.read_csv("../../개인 필기/pd/TSLA.csv")
tsla_df.Date = pd.to_datetime(tsla_df.Date)
tsla_df


# 1) High / Low /Close / Volume 열에 대해 최대/최소/평균/중위수 출력

tsla_df[['High', 'Low', 'Close', 'Volume']].agg(['max', 'min', 'mean', 'median'])


# 2) Open열과 Close 열 값의 차이가 가장 큰 날짜는 언제인가?

tsla_df['Diff'] = tsla_df.Open - tsla_df.Close

tsla_df.Date[tsla_df.Diff == tsla_df.Diff.max()]


# 3) Volume(거래량)이 가장 높았던 날에는 주가가 올랐을까 내렸을까?

max_vol_df = tsla_df[tsla_df.Volume == tsla_df.Volume.max()]

max_vol_df.Open < max_vol_df.Close


# 4) 3,4,5월 중에서 평균적으로 종가(Close)가 가장 높은 달은?

tsla_df.groupby(tsla_df.Date.dt.strftime('%Y년 %m월')).Close.mean().idxmax() # 날짜를 인덱스로 하는 평균값 시리즈에서 최댓값의 인덱스


# 5) 종가 및 거래량 데이터를 활용하여 시각화(선그래프) 출력

fig, axes1 = plt.subplots(figsize=(12, 6))

axes1.set_xlabel('Date') # x축 날짜
axes1.set_ylabel('Close', color='blue') # y축1 종가
axes1.plot(tsla_df.Date, tsla_df.Close, color='blue', label='종가')

axes2 = axes1.twinx()
axes2.set_ylabel('Volume', color='red') # y축2 거래량
axes2.plot(tsla_df.Date, tsla_df.Volume, color='red', label='거래량')


# 6) 주가가 오른 날이 많았을까? 내린 날이 많았을까?

(tsla_df.Open < tsla_df.Close).value_counts()

# False: 34건, True: 29건이고, False가 더 많으므로 오른날이 더 적다.


# 2. 인구

df = pd.read_csv('../../개인 필기/pd/gapminder.tsv', sep='\t')
df


# 1) 대륙별로 기대수명이 평균적으로 가장 높은 대륙은?

df.groupby('continent').lifeExp.mean().idxmax()


# 2) 대륙 별 국가의 개수를 출력하시오

df.groupby('continent').country.count()


# 3) 기대수명이 가장 높은 국가는?

df.country[df.lifeExp == df.lifeExp.max()]


# 4) 1952~2007년까지 gdp가 가장 크게 증가한 나라는? 증가율은?

y1952 = df[df.year == 1952].set_index('country').gdpPercap
y2007 = df[df.year == 2007].set_index('country').gdpPercap
increased_amount = y2007 - y1952

ctr = increased_amount.idxmax() # 국가
ser = increased_amount / y1952 * 100
ser[ser.index == ctr] # 수치


# 5) 1952~2007년까지 pop가 가장 크게 증가한 나라는? 증가율은?

y1952 = df[df.year == 1952].set_index('country')['pop']
y2007 = df[df.year == 2007].set_index('country')['pop']
increased_amount = y2007 - y1952

ctr = increased_amount.idxmax() # 국가
ser = increased_amount / y1952 * 100
ser[ser.index == ctr] # 수치
```
