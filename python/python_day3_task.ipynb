```python
# 1. 표준 입력으로 문자열 여러 개와 숫자(실수) 여러 개가 두 줄로 입력됩니다.
# 입력된 첫 번째 줄은 키, 두 번째 줄은 값으로 하여 딕셔너리를 생성한 뒤 딕셔너리를 출력하는 프로그램을
# 만드세요.
# input().split()의 결과를 변수 한 개에 저장하면 리스트로 저장됩니다.

# 예)
# 입력
# health health_regen mana mana_regen
# 575.6 1.7 338.8 1.63
# 결과
# {'health': 575.6, 'health_regen': 1.7, 'mana': 338.8, 'mana_regen': 1.63}

# 입력
# health mana melee attack_speed magic_resistance
# 573.6 308.8 600 0.625 35.7
# 결과
# {'health': 573.6, 'mana': 308.8, 'melee': 600.0, 'attack_speed': 0.625, 'magic_resistance': 35.7}

str_lst = input("문자열 입력: ").split()
num_lst = input("숫자 입력: ").split()
dic = {}

for i in range(len(str_lst)):
    dic[str_lst[i]] = num_lst[i]

print(dic)
# 2. 표준 입력으로 문자열 두 개가 각 줄에 입력됩니다(문자열의 길이는 정해져 있지 않음).
# 첫 번째 문자열에서 인덱스가 홀수인 문자와 두 번째 문자열에서 인덱스가 짝수인 문자를 연결하여
# 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다). 연결 순서는 첫 번째 문자열, 두 번째 문자열 순입니다. 그리고 0은 짝수로 처리합니다.

# 예
# 입력
# python
# python
# 결과
# yhnpto
# 입력
# apple
# strawberry
# 결과
# plsrwer

str_lst = input()
str_lst2 = input()

str_lst = str_lst[1::2]
str_lst2 = str_lst2[::2]

result = "".join(str_lst + str_lst2)
print(result)
# 3. 점수 구간에 해당하는 학점이 아래와 같이 정의되어 있다.
# 사용자로부터 score를 입력받아 학점을 출력하라.

# 점수	학점
# 81~100	A
# 61~80	B
# 41~60	C
# 21~40	D
# 0~20	E
# >> score: 83
# grade is A

score = int(input())

if 81 <= score <= 100:
    score = 'A'
elif 61 <= score <= 80:
    score = 'B'
elif 41 <= score <= 60:
    score = 'C'
elif 21 <= score <= 40:
    score = 'B'
elif 0 <= score <= 20:
    score = 'E'

print("grade is {s}".format(s = score))
# 4. for문을 사용해 1부터 100까지의 숫자를 출력해 보자.

for i in range(1, 101):
    print(i, end=(", "))
# 5.
# A 학급에 총 10명의 학생이 있다. 이 학생들의 중간고사 점수는 다음과 같다.
# [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]
# for문을 사용하여 A 학급의 평균 점수를 구해 보자.

class_A = [70, 60, 55, 75, 95, 90, 80, 80, 85, 100]

avg = sum(class_A) / len(class_A)
print(avg)
# 6.
# 사용자로부터 달러, 엔, 유로, 또는 위안 금액을 입력받은 후 이를 원으로 변환하는 프로그램을 작성하라.
# 각 통화별 환율은 다음과 같다. 사용자는 100 달러, 1000 엔, 13 유로, 100 위안과 같이 금액과 통화명 사이에
#  공백을 넣어 입력한다고 가정한다.

# 통화명	환율
# 달러	1367
# 엔	1.196
# 유로	1368
# 위안	181
# >> 입력: 100 달러
# 136700.00 원

ecr = {'달러' : 1367,
         '엔' : 1.196,
       '유로' : 1368,
       '위안' : 181
       }

money = input().split()
result = 0

for k, v in ecr.items():
    if k == money[1]:
        result = v * int(money[0])
        print(result, "원")
        break
# 7.
# 사용자로부터 세 개의 숫자를 입력 받은 후 가장 큰 숫자를 출력하라.

# >> input number1: 10
# >> input number2: 9
# >> input number3: 20
# 20

number1, number2, number3 = map(int, (input("숫자 세개 입력: ").split()))
nums = [number1, number2, number3]
nums.sort()

result = len(nums) - 1
print(nums[result])
# 8.
# 파일 이름이 저장된 리스트에서 확장자를 제거하고 파일 이름만 화면에 출력하라. (힌트: split() 메서드)

# 리스트 = ['hello.py', 'ex01.py', 'intro.hwp']
# hello
# ex01
# intro

files = ['hello.py', 'ex01.py', 'intro.hwp']

for i in files:
    i = i.split(".")
    print(i[0])
# 9.
# 파일 이름이 저장된 리스트에서 확장자가 .h인 파일 이름을 출력하라.

# 리스트 = ['intra.h', 'intra.c', 'define.h', 'run.py']
# intra.h
# define.h

files = ['intra.h', 'intra.c', 'define.h', 'run.py']

for i in files:
    if i.find(".h") != -1:
        print(i)
# 10.
# 파일 이름이 저장된 리스트에서 확장자가 .h나 .c인 파일을 화면에 출력하라.

# 리스트 = ['intra.h', 'intra.c', 'define.h', 'run.py']
# intra.h
# intra.c
# define.h

files = ['intra.h', 'intra.c', 'define.h', 'run.py']

for i in files:
    if i.find(".h") != -1 or i.find(".c") != -1:
        print(i)
# 11. 자판기(pro, 커피 한 잔에 300원이라 가정, 초기 커피는 10개)
# 돈을 넣어 주세요: 500
# 거스름돈 200를 주고 커피를 줍니다.
# 돈을 넣어 주세요: 300
# 커피를 줍니다.
# 돈을 넣어 주세요: 100
# 돈을 다시 돌려주고 커피를 주지 않습니다.
# 돈을 넣어 주세요: 0
# 종료합니다

coffee_count = 10
coffee_price = 300

while True:
    if coffee_count == 0:
        print("재고 소진")
        break
    else:
        print("재고: %d개" % coffee_count)

    pay = int(input("돈을 넣어 주세요: "))
    change = 0

    if pay >= coffee_price:
        coffee_count -= 1
        change = pay - coffee_price
        if change == 0:
            print("커피를 줍니다.")
        else:
            print("거스름돈 %d를 주고 커피를 줍니다." % change)
    elif pay == 0:
        print("종료합니다")
        break
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
# 12. while문을 사용하여 다음과 같이 별(*)을 표시하는 프로그램을 작성해 보자.
# *
# **
# ***
# ****
# *****

count = 1

while count < 6:
    print("*" * count)
    count += 1
# 13
# 아래 표에서 하나의 행을 하나의 리스트로, 총 2개의 리스트를 갖는 이차원 리스트 apart를 정의하라.

# 101호	102호	103호
# 201호	202호	203호

aprt = [['101호', '102호', '103호'], ['201호', '202호', '203호']]
print("   ".join(aprt[0]))
print("   ".join(aprt[1]))
# 14
# 아래 표에서 하나의 열을 하나의 리스트로, 총 2개의 리스트를 갖는 이차원 리스트 stock을 정의하라.

# 시가	종가
# 100	80
# 200	210
# 300	330

stock = [['시가', 100, 200, 300], ['종가', 80, 210, 330]]
for i in (range(4)):
    print(stock[0][i], " ", stock[1][i])
# 15
# 아래 표를 stock 이름의 딕셔너리로 표현하라. 시가를 key로 저장하고, 나머지 같은 열의 데이터를 리스트로 저장해서 value로 저장한다. 종가를 key로 저장하고 나머지 같은 열의 데이터를 리스트로 저장해서 value로 저장한다.

# 시가	종가
# 100	80
# 200	210
# 300	330

siga = [100, 200 ,300]
jonga = [80, 210, 330]
stock = {}

print("시가", " ", "종가")
for i in range(3):
    stock[siga[i]] = jonga[i]
    print(siga[i], "  ", jonga[i])
# 16
# 아래 표를 stock 이름의 딕셔너리로 표현하라. 날짜를 key로 저장하고, 나머지 같은 행의 데이터를 리스트로 저장해서 value로 저장한다. 첫 열이 날짜이다.

# 10/10	80	110	70	90
# 10/11	210	230	190	200

key_date = ['10/10', '10/11']
value_list = [80, 110, 70, 90]
value_list2 = [210, 230, 190, 200]
stock = {}

stock[key_date[0]] = value_list
stock[key_date[1]] = value_list2
print(stock)
# 17
# 리스트에 저장된 데이터를 아래와 같이 출력하라.

# apart = [ [101, 102], [201, 202], [301, 302] ]
# 101 호
# 102 호
# 201 호
# 202 호
# 301 호
# 302 호

apart = [ [101, 102], [201, 202], [301, 302] ]

for i in range(3):
    print("{a}호\n{b}호".format(a = apart[i][0], b = apart[i][1]))
# 18
# 리스트에 저장된 데이터를 아래와 같이 출력하라.

# apart = [ [101, 102], [201, 202], [301, 302] ]
# 301 호
# 302 호
# 201 호
# 202 호
# 101 호
# 102 호

# 첫번째 풀이
apart = [ [101, 102], [201, 202], [301, 302] ]

for i in range(len(apart) - 1, -1, -1):
    print("{a}호\n{b}호".format(a = apart[i][0], b = apart[i][1]))

# 두번째 풀이
apart = [ [101, 102], [201, 202], [301, 302] ]

r_r = list(range(3))
r_r.reverse()

for i in r_r:
    print("{a}호\n{b}호".format(a = apart[i][0], b = apart[i][1]))
# 19
# 리스트에 저장된 데이터를 아래와 같이 출력하라.

# apart = [ [101, 102], [201, 202], [301, 302] ]
# 302 호
# 301 호
# 202 호
# 201 호
# 102 호
# 101 호

apart = [ [101, 102], [201, 202], [301, 302] ]

for i in range(len(apart) - 1, -1, -1):
    print("{a}호\n{b}호".format(a = apart[i][1], b = apart[i][0]))
# 20
# 리스트에 저장된 데이터를 아래와 같이 출력하라.

# apart = [ [101, 102], [201, 202], [301, 302] ]
# 101 호
# -----
# 102 호
# -----
# 201 호
# -----
# 202 호
# -----
# 301 호
# -----
# 302 호

apart = [ [101, 102], [201, 202], [301, 302] ]

for i in range(3):
    print("{a}호\n-----\n{b}호".format(a = apart[i][0], b = apart[i][1]))
    if i != 2:
        print('-----')
# 21
# 리스트에 저장된 데이터를 아래와 같이 출력하라.

# apart = [ [101, 102], [201, 202], [301, 302] ]
# 101 호
# 102 호
# -----
# 201 호
# 202 호
# -----
# 301 호
# 302 호

apart = [ [101, 102], [201, 202], [301, 302] ]

for i in range(3):
    print("{a}호\n{b}호".format(a = apart[i][0], b = apart[i][1]))
    if i != 2:
        print('-----')
# 22
# 리스트에 저장된 데이터를 아래와 같이 출력하라.

# apart = [ [101, 102], [201, 202], [301, 302] ]
# 101 호
# 102 호
# 201 호
# 202 호
# 301 호
# 302 호
# -----

apart = [ [101, 102], [201, 202], [301, 302] ]

for i in range(3):
    print("{a}호\n{b}호".format(a = apart[i][0], b = apart[i][1]))
    if i == 2:
        print('-----')
```

```python
# 주어진 문자열(공백 없이 쉼표로 구분되어 있음)을 가지고 아래 문제에 대한 프로그램을 작성하세요.

names = "이유덕,이재영,권종표,이재영,박민호,강상희,이재영,김지완,최승혁,이성연,박영서,박민호,전경헌,송정환,김재성,이유덕,전경헌"
names = list(names.split(','))
# print(names)

# 김씨와 이씨는 각각 몇 명 인가요?

kim_count = 0
lee_kount = 0

for i in names:
    if i[0] == '김':
        kim_count += 1
    elif i[0] == '이':
        lee_kount += 1
print("김씨: %d명\n이씨: %d명" % (kim_count, lee_kount))

# "이재영"이란 이름이 몇 번 반복되나요?

repeat_count = -1

for i in names:
    if i == '이재영':
        repeat_count += 1
print("%d번 반복" % repeat_count)

# 중복을 제거한 이름을 출력하세요.

names = list(set(names))
print(names)

# 중복을 제거한 이름을 오름차순으로 정렬하여 출력하세요.

names.sort()
print(names)
```
