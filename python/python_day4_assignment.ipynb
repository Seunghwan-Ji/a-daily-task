```python
# "비트코인" 문자열을 화면에 출력하는 print_coin() 함수를 정의하라.

def print_coin():
    print("비트코인")
# 2
# 1 번에서 정의한 함수를 호출하라.

print_coin()
# 3
# 1 번에서 정의한 함수를 100번호출하라.

call = 0 # 호출 횟수 저장

for i in range(100):
    print_coin()
    call += 1
print(call)
# 4
# "비트코인" 문자열을 100번 화면에 출력하는 print_coins() 함수를 정의하라.
# 한 라인에 하나씩 "비트코인" 문자열을 출력한다.

call = 0

def print_coin():
    for i in range(100):
        print("비트코인")
        global call
        call += 1

print_coin()
print(call)
# 5
# 아래의 에러가 발생하는 이유에 대해 설명하라.

# hello()
# def hello():
#     print("Hi")
# NameError: name 'hello' is not defined

# -> 함수 정의를 먼저 하고나서 호출해야 합니다.
# 6
# 아래 코드의 실행 결과를 예측하라.

def message() :
    print("A")
    print("B")

message()
print("C")
message()

# -> (A 출력, B 출력) -> C 출력 -> (A 출력, B 출력)
# 7
# 아래 코드의 실행 결과를 예측하라.

print("A")

def message() :
    print("B")

print("C")
message()

# -> A 출력 -> C 출력 -> B 출력
# 8
# 아래 코드의 실행 결과를 예측하라.

print("A")
def message1() :
    print("B")
print("C")
def message2() :
    print("D")
message1()
print("E")
message2()

# -> A 출력 -> C 출력 -> B 출력 -> E 출력 -> D 출력
# 9
# 아래 코드의 실행 결과를 예측하라.

def message1():
    print("A")

def message2():
    print("B")
    message1()

message2()

# -> B 출력 -> A 출력
# 10
# 아래 코드의 실행 결과를 예측하라.

def message1():
    print("A")

def message2():
    print("B")


def message3():
    for i in range (3) :
        message2()
        print("C")
    message1()

message3()

# -> (B 출력, C 출력) 3번 반복 -> A 출력
# 11
# 함수의 호출 결과를 예측하라.

def 함수(문자열) :
    print(문자열)

함수("안녕")
함수("Hi")

# -> "안녕" 출력 -> "Hi" 출력
# 12
# 함수의 호출 결과를 예측하라.

def 함수(a, b) :
    print(a + b)

함수(3, 4)
함수(7, 8)

# -> 7 출력 -> 15 출력
# 13
# 아래와 같은 에러가 발생하는 원인을 설명하라.

# def 함수(문자열) :
#     print(문자열)
# 함수()
# TypeError: 함수() missing 1 required positional argument: '문자열'

# -> 함수가 문자열 변수(파라미터)를 받아야 하는데, 호출할 때 전달할 값(아규먼트)을 주지 않았습니다.
# 14
# 아래와 같은 에러가 발생하는 원인을 설명하라.

# def 함수(a, b) :
#     print(a + b)

# 함수("안녕", 3)
# TypeError: must be str, not int

# -> 함수는 호출되었지만 실행문에서는 문자열과 숫자는 더해질 수 없기때문에 에러가 발생합니다.
# 15
# 하나의 문자를 입력받아 문자열 끝에 ":D" 스마일 문자열을
# 이어 붙여 출력하는 print_with_smile 함수를 정의하라.

def print_with_smile(prmt):
    result = prmt + ":D"
    print(result)

agm = input("문자열 입력: ")
# 16
# 015에서 정의한 함수를 호출하라. 파라미터는 "안녕하세요"로 입력하라.

agm = input("문자열 입력: ")
print_with_smile(agm)
# 17
# 현재 가격을 입력 받아 상한가 (30%)를 출력하는 print_upper_price 함수를 정의하라.

def print_upper_price(v):
    print(v + (0.3 * v))

price = int(input("가격 입력: "))
print_upper_price(price)
# 18
# 두 개의 숫자를 입력받아 두 수의 합을 출력하는 print_sum 함수를 정의하라.

def print_sum(num1, num2):
    print(num1 + num2)

n1, n2 = map(int, input("두 수 입력: ").split())
print_sum(n1, n2)
# 19
# 두 개의 숫자를 입력받아 합/차/곱/나눗셈을 출력하는 print_arithmetic_operation 함수를 작성하라.

# print_arithmetic_operation(3, 4)
# 3 + 4 = 7
# 3 - 4 = -1
# 3 * 4 = 12
# 3 / 4 = 0.75

def print_arithmetic_operation(num1, num2):
    print(num1 + num2)
    print(num1 - num2)
    print(num1 * num2)
    print(num1 / num2)

n1, n2 = map(int, input("두 수 입력: ").split())
print_arithmetic_operation(n1, n2)
# 20
# 세 개의 숫자를 입력받아 가장 큰수를 출력하는 print_max 함수를 정의하라.
# 단 if 문을 사용해서 수를 비교하라.

def print_max(num1, num2, num3):
    lst = [num1, num2, num3]
    cv = 0
    for i in lst:
        if i > cv:
            cv = i
    print(cv)

n1, n2, n3 = map(int, input("세 수 입력: ").split())
print_max(n1, n2, n3)
```
