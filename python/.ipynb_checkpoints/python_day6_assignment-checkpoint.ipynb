```python
# 1
# "test.txt"라는 파일에 "Life is too short" 문자열을 저장한 후
# 다시 그 파일을 읽어서 출력하는 프로그램을 작성해 보자.

with open("test.txt", "w") as f:
    f.write("Life is too short")

with open("test.txt", "r") as f:
    print(f.read())


# 2
# 다음과 같은 내용을 지닌 파일 test.txt가 있다.
# 이 파일의 내용 중 "java"라는 문자열을 "python"으로 바꾸어서 저장해 보자.

# Life is too short
# you need java
# ※ replace 함수를 사용해 보자

with open("test.txt", "a") as f:
    f.write("\nyou need java".replace("java", "python"))


# 3
# 369 게임은 숫자에 3, 6, 9가 들어가면 박수를 치고, 나머지 숫자는 그대로 이야기하는 게임입니다.
# 369 게임을 시뮬레이션하는 game_369 함수를 정의하라. 함수는 숫자의 리스트를 입력받는다.
# 리스트에는 0 ~ 9 까지의 숫자가 들어있다. 입력된 숫자에 3/6/9가 포함되면 "짝" 문자열을
# 그렇지 않을 경우 숫자를 그대로 리스트에 저장한다. * 함수는 변환된 리스트를 반환한다.

# [1, 2, "짝", 4, 5, "짝", 7, 8, "짝"]

def game_369(num_lst):
    str_lst = []
    result = []
    for i in num_lst:
        str_lst.append(str(i)) # find함수를 쓰기 위해 타입 변환
    # print(str_lst)
    for j in str_lst:
        if j.find('3') != -1 or j.find('6') != -1 or j.find('9') != -1:
            result.append("짝")
        else:
            result.append(int(j)) # 다시 초기타입으로 변환해 리스트로 보냅니다.
    return result

print(game_369( [1, 2, 3, 4, 5, 6, 7, 8, 9] ))


# 4
# 3번에 정의한 함수를 0~99 까지 정상동작하도록 수정하라.

# game_369( [31, 32, 33, 34] )
# ["짝", "짝", "짝", "짝"]

print(game_369(range(100)))
print(game_369(range(31, 35)))


# 5
# 아래 코드는 임의의 수를 생성하는 코드이다. randint(1, 10)은 1부터 10사이의 수를 임의로 생성한다.
# 코드를 실행할 때마다 다른 숫자가 출력된다.

import random
# print(random.randint(1, 10))

# 중복되는 수가 없이 6개의 로또 번호를 생성하는 make_lotto_without_replacement 함수를 정의하라.

def make_lotto_without_replacement(count):
    lotto_list = []

    for i in range(count):
        lotto = []
        while len(lotto) != 6: # 로또 1줄의 길이가 6이 될때까지 수행합니다.
            nums = random.choice(range(1, 46)) # 1~45까지의 요소중 하나를 초이스
            lotto.append(nums)
            set(lotto) # 중복 요소 제거
        lotto.sort() # 오름차순 정렬
        lotto_list.append(lotto) # 1줄이 완성되어 리스트로 보내집니다.
        print("{0}. {1}".format(i + 1, lotto)) # 입력받은 줄만큼 출력합니다.
    return lotto_list

while True:
    try:
        lotto_count = int(input("로또 줄 수 입력: "))
    except:
        print("숫자만 입력하세요.(소수 불가능)")
        continue
    else:
        if lotto_count <= 0:
            print("0보다 큰 숫자만 입력하세요.")
            continue
        else:
            make_lotto_without_replacement(lotto_count)
            break


# 6
# 표준 입력으로 숫자.확장자 형식으로 된 파일 이름 여러 개가 입력됩니다.
# 파일 이름이 숫자 3개이면서 앞에 0이 들어가는 형식으로 출력되게 만드세요.
# 예를 들어 1.png는 001.png, 99.docx는 099.docx, 100.xlsx는 100.xlsx처럼 출력되어야 합니다.
# 그리고 람다 표현식을 사용해야 하며 출력 결과는 리스트 형태라야 합니다.
# 람다 표현식에서 파일명을 처리할 때는 문자열 포매팅과 문자열 메서드를 활용하세요.

# 예
# 입력
# 1.jpg 10.png 11.png 2.jpg 3.png
# 결과
# ['001.jpg', '010.png', '011.png', '002.jpg', '003.png']
# 입력
# 97.xlsx 98.docx 99.docx 100.xlsx 101.docx 102.docx
# 결과
# ['097.xlsx', '098.docx', '099.docx', '100.xlsx', '101.docx', '102.docx']

file_names = input("파일명 입력: ").split()
mdf = []
# print(file_names)

for i in file_names:
    name, ext = i.split('.')[0], i.split('.')[1] # 파일명과 확장자를 분리합니다.
    name_len = len(name)
    if name_len == 1:
        mdf.append('00' + name + '.' + ext)
    elif name_len == 2:
        mdf.append('0' + name + '.' + ext)
    else:
        mdf.append(i)
print(mdf)


# 7  DashInsert 함수
# DashInsert 함수는 숫자로 구성된 문자열을 입력받은 뒤 문자열 안에서 홀수가 연속되면
# 두 수 사이에 - 를 추가하고, 짝수가 연속되면 * 를 추가하는 기능을 갖고 있다.
# DashInsert 함수를 완성하시오.

# 입력 예시: 4546793
# 출력 예시: 454*67-9-3

work = True

def DashInsert():
    global work
    while work:
        past_len = len(str_num) # 초기 길이를 저장합니다.
        for i in range(len(str_num) - 1):
            try: # 특수문자 타입변환 오류를 방지합니다.
                if int(str_num[i]) % 2 == 1 and int(str_num[i + 1]) % 2 == 1:
                    str_num.insert(i + 1, '-') # 기존 리스트에서 i와 i+1 요소 사이에 특수문자 요소 추가
                elif int(str_num[i]) % 2 == 0 and int(str_num[i + 1]) % 2 == 0:
                    str_num.insert(i + 1, '*')
            except: # 오류가 났다면 i가 특수 문자이므로 스킵하고 다음 반복 수행
                continue
        if len(str_num) == past_len: # str_num리스트에 값을 추가할 때마다
            work = False             # 길이가 길어지므로 더이상 길어지지
            print("작업 완료")       # 않을때까지 비교합니다.

    # while 목적은 반복문 수행중에 참조중인 리스트에 값이 추가되면
    # 한 번 실행만으론 마지막 요소까지 접근할 수 없기 때문입니다.
    # 다음 실행에서 값이 추가되지 않을때까지 위로 보냅니다.
    # 메모리 스택 쌓임 방지를 위해 재귀 대신 while 사용

str_num = list(input("숫자 입력: "))
DashInsert()
print("".join(str_num))


# 8 문자열 압축하기
# 문자열을 입력받아 같은 문자가 연속적으로 반복되는 경우에
# 그 반복 횟수를 표시해 문자열을 압축하여 표시하시오.

# 입력 예시: aaabbcccccca
# 출력 예시: a3b2c6a1

strs = input("문자 입력: ")

target = ""
count = 0
target_count = {}
record = []

# 중복되는 키가 들어오면 기록해뒀던 밸류값이 수정되기 때문에
# 타겟이 바뀔때마다 이 함수를 호출해 다른 리스트에 담고
# 다시 딕셔너리를 초기화 합니다.
def recording():
    global target_count
    for k, v in target_count.items():
        record.append(k + v) # 문자열 + 문자열("타겟" + "반복횟수")을 리스트에 추가합니다.
        target_count = {} # 딕셔너리 초기화

for i in strs:
    if i == target:
        count += 1
    else:
        if target_count:
            recording()
        target = i # 타겟 변경
        count = 0 # 반복횟수 초기화
        count += 1 # 반복횟수
    target_count[target] = str(count) # 나중에 문자열 타겟과 합치기 위해 타입 변환
recording() # 반복문 마치고 마지막 기록까지 저장
print("".join(record))
```

```python
# 6번 문제 람다 표현식으로 풀이

# 람다 표현식을 사용해야 하며 출력 결과는 리스트 형태라야 합니다.
# 람다 표현식에서 파일명을 처리할 때는 문자열 포매팅과 문자열 메서드를 활용하세요.

# 예
# 입력
# 1.jpg 10.png 11.png 2.jpg 3.png
# 결과
# ['001.jpg', '010.png', '011.png', '002.jpg', '003.png']
# 입력
# 97.xlsx 98.docx 99.docx 100.xlsx 101.docx 102.docx
# 결과
# ['097.xlsx', '098.docx', '099.docx', '100.xlsx', '101.docx', '102.docx']

files = input("파일 입력: ").split() # 스플릿해서 files 리스트에 저장

lam_f = lambda f : "{:03}.{}".format(int(f.split(".")[0]), f.split(".")[1]) # 람다 함수 정의

result = list(map(lam_f, files)) # files 리스트를 정의된 람다 함수로 맵핑하고 리스트로 변환

print(result)
```

```python
# 7번 문제 재풀이

def DashInsert(strs):
    result = []
    jjak = 0
    hol = 0
    for i in strs:
        if int(i) % 2 == 0: # 짝수면
            jjak += 1 # 짝 스택 +1
            hol = 0 # 홀 스택 초기화
            if jjak == 2: # 짝 스택이 2까지 쌓이면
                result.append("*") # 리스트에 "*" 추가
                jjak -= 1 # 짝 스택 -1
        else: # 홀수면
            hol += 1 # 홀 스택 +1
            jjak = 0 # 짝 스택 초기화
            if hol == 2: # 홀 스택이 2까지 쌓이면
                result.append("-") # 리스트에 "-" 추가
                hol -= 1 # 홀 스택 -1
        result.append(i) # 리스트에 숫자 추가
    return "".join(result)

print(DashInsert("4546793"))
```

```python
# 8번 문제 재풀이
# 문자열을 입력받아 같은 문자가 연속적으로 반복되는 경우에
# 그 반복 횟수를 표시해 문자열을 압축하여 표시하시오.

# 입력 예시: aaabbcccccca
# 출력 예시: a3b2c6a1

strs = input("문자 입력: ")

target = ""
count = 0
target_count = {}
record = []

def recording():
    k = list(target_count.keys())[0] # 딕셔너리의 키 값만 추출하는 과정입니다.
    record.append(k + target_count[k]) # record 리스트에 "키" + "밸류" 를 저장합니다.
    return "".join(record) # record 리스트의 요소들을 연결하여 새로운 문자열을 생성 후 반환합니다.

for i in strs:
    if i == target:
        count += 1 # 타겟의 반복횟수 업데이트
    else:
        if target_count:
            recording() # 타겟이 바뀌기 전 타겟의 정보(키(타겟), 밸류(반복횟수))를 기록하는 함수 입니다.
        target_count = {} # 다음 타겟의 정보를 기록하기 위해 딕셔너리를 초기화 합니다.
        target = i # 타겟 변경
        count = 0 # 반복횟수 초기화
        count += 1 # 새 타겟의 반복횟수 업데이트
    target_count[target] = str(count) # 타겟, 반복횟수를 키:밸류 한 쌍으로 저장합니다. recording 함수에서 문자열 타겟과 합치기 위해 타입 변환
result = recording() # 반복문 마치고 마지막 기록까지 저장
print(result)
```
