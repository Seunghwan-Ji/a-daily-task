```python
# 21
# 입력된 문자열을 역순으로 출력하는 print_reverse 함수를 정의하라.

# print_reverse("python")
# nohtyp

def print_reverse(input_str):
    print(input_str)

input_str = "".join(reversed(input("문자열 입력: ")))
print_reverse(input_str)


# 22
# 성적 리스트를 입력 받아 평균을 출력하는 print_score 함수를 정의하라.
# print_score ([1, 2, 3])
# 2.0

def print_score(score_lst):
    print(sum(score_lst) / len(score_lst))

input_score = list(map(int, input("성적 입력: ").split()))
print_score(input_score)


# 23
# 하나의 리스트를 입력받아 짝수만 화면에 출력하는 print_even 함수를 정의하라.

# print_even ([1, 3, 2, 10, 12, 11, 15])
# 2
# 10
# 12

# 입력: 1 3 2 10 12 11 15

# 첫번째 풀이
def print_even(fil):
    for i in fil: # filter도 for문에서 참조할 수 있는거 같다.
        print(i)

input_number = list(map(int, input("숫자 입력: ").split()))
print_even(filter(lambda i : i % 2 == 0, input_number))

# 두번째 풀이
def print_even(num):
    return str(num)

print("\n".join(list(map(print_even, filter(lambda i : i % 2 == 0, list(map(int, input("숫자 입력: ").split())))))))


# 24
# 하나의 딕셔너리를 입력받아 딕셔너리의 key 값을 화면에 출력하는 print_keys 함수를 정의하라.

# print_keys ({"이름":"김말똥", "나이":30, "성별":0})
# 이름
# 나이
# 성별

# 입력: 이름 김말똥 나이 30 성별 0

def print_keys(dic):
    key_lst = list(dic.keys()) # 키 들만 빼와서 저장
    print("\n".join(key_lst)) # 줄바꿈 문자로 이어붙여 새로운 문자열 생성

input_profile = input("개인정보 입력: ").split(" ")
k_lst = input_profile[::2] # 키 저장
v_lst = input_profile[1::2] # 밸류 저장
profile = dict(zip(k_lst, v_lst)) # 인덱스 번호끼리 튜플로 묶고 딕셔너리로 변환
print_keys(profile)

# 질문
# dict(map(lambda k, v : ([k, v]), k_lst, v_lst)) # 오류 안남 why?
# dict(map(lambda k, v : [(k, v)], k_lst, v_lst)) # 오류 남 why?
# dict(map(lambda k, v : (k, v), k_lst, v_lst)) # 오류 안남 why?
# dict(map(lambda k, v : [k, v], k_lst, v_lst)) # 오류 안남 why?


# 25
# my_dict에는 날짜를 키값으로 OHLC가 리스트로 저장돼 있다.

my_dict = {"10/26" : [100, 130, 100, 100],
           "10/27" : [10, 12, 10, 11]}

# my_dict와 날짜 키값을 입력받아 OHLC 리스트를 출력하는 print_value_by_key 함수를 정의하라.

# print_value_by_key  (my_dict, "10/26")
# [100, 130, 100, 100]

def print_value_by_key(dic, k):
    print(dic[k])

print_value_by_key(my_dict, "10/26")


# 26
# 입력 문자열을 한 줄에 다섯글자씩 출력하는 print_5xn(string) 함수를 작성하라.

# print_5xn("아이엠어보이유알어걸")
# 아이엠어보
# 이유알어걸

def print_5xn(string):
    count = 0
    while 1:
        str5 = string[count:count+5] # string[0:5], string[5:10]...
        count += 5 # 다음 글자 시작 번호
        if str5:
            print(str5)
        else:
            break # str5이 공백이면 중지

print_5xn("아이엠어보이유알어걸")


# 27
# 문자열과 한줄에 출력될 글자 수를 입력을 받아 한 줄에 입력된 글자 수만큼 출력하는 print_mxn(string) 함수를 작성하라.

# printmxn("아이엠어보이유알어걸", 3)
# 아이엠
# 어보이
# 유알어
# 걸

# 위 문제 함수 로직과 동일
def printmxn(string, num):
    count = 0
    while 1:
        str3 = string[count:count+3]
        count += 3
        if str3:
            print(str3)
        else:
            break

printmxn("아이엠어보이유알어걸", 3)


# 28
# 연봉을 입력받아 월급을 계산하는 calc_monthly_salary(annual_salary) 함수를 정의하라.
# 회사는 연봉을 12개월로 나누어 분할 지급하며, 이 때 1원 미만은 버림한다.

# calc_monthly_salary(12000000)
# 1000000

def calc_monthly_salary(annual_salary):
    print(int(annual_salary / 12))

calc_monthly_salary(12000000)


# 29
# 아래 코드의 실행 결과를 예측하라.

def my_print (a, b) :
    print("왼쪽:", a)
    print("오른쪽:", b)

my_print(a=100, b=200)

# 왼쪽: 100
# 오른쪽: 200


# 30
# 아래 코드의 실행 결과를 예측하라.

def my_print (a, b) :
    print("왼쪽:", a)
    print("오른쪽:", b)

my_print(b=100, a=200)

# 왼쪽: 200
# 오른쪽: 100


# 31
# 아래 코드를 실행한 결과를 예상하라.

def n_plus_1(n):
    result = n + 1

n_plus_1(3)
# print(result) # result는 함수 내 에서만 선언된 지역변수이기 때문에
              # 출력하기 전에 값을 읽지 못하고 오류가 발생합니다.


# 32
# 문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 정의하라.

# make_url("naver")
# www.naver.com

def make_url(string):
    return "www." + string + ".com"

make_url("naver")


# 33
# 문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하라.

# make_list("abcd")
# ['a', 'b', 'c', 'd']

def make_list(string):
    return list(string[:])

make_list("abcd")


# 34
# 숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하라.

# pickup_even([3, 4, 5, 6, 7, 8])
# [4, 6, 8]

def pickup_even(lst):
    return list(filter(lambda i : i % 2 == 0, lst))

pickup_even([3, 4, 5, 6, 7, 8])


# 35
# 콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하라.

# convert_int("1,234,567")
# 1234567

def convert_int(string):
    lst = string[:]
    convert = []
    for i in lst:
        try:
            int(i)
        except:
            continue
        else:
            convert.append(i)
    return "".join(convert)

convert_int("1,234,567")


# 36
# 아래 코드의 실행 결과를 예측하라.

def 함수(num) :
    return num + 4

a = 함수(10) # 14 반환
b = 함수(a) # 18 반환
c = 함수(b) # 22 반환
print(c) # 22 출력


# 37
# 아래 코드의 실행 결과를 예측하라.

def 함수(num) :
    return num + 4

c = 함수(함수(함수(10))) # 함수(함수(14)) -> 함수(18) -> 22
print(c) # 22 출력


# 38
# 아래 코드의 실행 결과를 예측하라.

def 함수1(num) :
    return num + 4

def 함수2(num) :
    return num * 10

a = 함수1(10) # 14 반환
c = 함수2(a) # 140 반환
print(c) # 140 출력


# 39
# 아래 코드의 실행 결과를 예측하라.

def 함수1(num) :
    return num + 4

def 함수2(num) :
    num = num + 2
    return 함수1(num)

c = 함수2(10) # 16 반환
print(c) # 16 출력


# 40
# 아래 코드의 실행 결과를 예측하라.

def 함수0(num) : # 06
    return num * 2 # 07

def 함수1(num) : # 04
    return 함수0(num + 2) # 05

def 함수2(num) : # 01
    num = num + 10 # 02
    return 함수1(num) # 03

c = 함수2(2) # 28 반환
print(c) # 28 출력
```
