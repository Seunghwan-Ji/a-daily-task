```python
# 1. 0부터 9999까지 8을 포함하지 않는 수는 총 몇개일까?
# 8, 108, 888, 9998 등은 8을 포함하고 있는 수입니다. 111, 299, 4 등과 같은 수는 8을 포함하지 않는 수 입니다.

nums = list(map(str, range(10000)))
# print(nums)

# 첫번째 풀이
eight_lst = []

for i in nums:
    if "8" in i:
        eight_lst.append(i)
print(len(nums) - len(eight_lst))

# 두번째 풀이
eight_lst = [i for i in nums if "8" in i]
print(len(nums) - len(eight_lst))


# 2.
# 컴퓨터가 1~100 숫자(정수 범위) 중 하나를 랜덤으로 정합니다. (이를 알려주지 않습니다.)
# 사용자는 이 숫자를 맞추어야 합니다.
# 입력한 숫자보다 정답이 크면 → "UP" 출력,
# 입력한 숫자보다 정답이 작으면 → "DOWN" 출력.
# 정답을 맞추면 → "정답"을 출력하고, 지금까지 숫자를 입력한 횟수를 알려줍니다.

# (예시)
# 컴퓨터가 1~100 중 랜덤 숫자 하나를 정합니다.
# 이 숫자를 맞춰주세요.
# 1~100 숫자 입력:50
# DOWN
# 1~100 숫자 입력:25
# UP
# 1~100 숫자 입력:38
# DOWN
# 1~100 숫자 입력:32
# UP
# 1~100 숫자 입력:35
# UP
# 1~100 숫자 입력:37
# DOWN
# 1~100 숫자 입력:36
# 정답입니다! 7회 만에 맞췄어요.

import random as rd

nansoo = rd.randint(1, 100)
# print(nansoo)
count = 0

while 1:
    num = int(input("1~100 정수 입력: "))
    count += 1
    if num < nansoo:
        print("UP")
        continue
    elif num > nansoo:
        print("DOWN")
        continue
    elif num == nansoo:
        print("정답")
        print(count)


# 3.직각삼각형의 밑변이 x, 높이가 y일 때 남은 변(대각선)의 길이를 구하는 함수를 만들어주세요.

def diagonal(x, y):
    return (x ** 2 / x) + (y ** 2 / y) # 루트x제곱 + 루트y제곱

base, height = list(map(float, input("밑변, 높이: ").split()))
diagonal(base, height)

# 4. 0~9까지의 문자로 된 숫자를 입력 받았을 때,
# 이 입력 값이 0~9까지의 숫자가 각각 한 번 씩만 사용된 것인지 확인하는 함수를 구하시오.

# sample inputs: 0123456789 01234 01234567890 6789012345 012322456789
# sample outputs: true false false true false

import re

def true_or_false(sample_inputs):
    for i in sample_inputs:
        res = re.match("[0-9]{10}$", i) # 숫자로된 10자리인지
        if res:
            if len(i) == len("".join(list(set(i)))): # 숫자 중복이 없는 10자리인지
                print("true")
            else:
                print("false")
        else:
            print("false")

sample_inputs = input("숫자 입력: ").split()
true_or_false(sample_inputs)

sample_inputs = input("숫자 입력: ").split()
true_or_false(sample_inputs)

# 두번째 풀이
data=input()
print('true' if len(data)==len(set(data))==10 else 'false')

# 5. 양의 정수만 입력으로 받고 그 수의 자릿수를 출력해보자. ex1) 3 > 1자리수, ex2) 649 > 3자리수 ....

while 1:
    try:
        input_num = int(input("양의 정수 입력: "))
    except:
        print("숫자만 입력하세요.(소수 불가능)")
        continue
    else:
        if input_num <= 0:
            print("0보다 큰 숫자만 입력하세요.")
            continue
        else:
            print("{0}자리수".format(len(str(input_num))))
            break


# 6. A사무실에는 특정일자의 출퇴근 시간이 기록된 거대한 로그파일이 있다고 한다.

# 파일의 형식은 다음과 같다. (한 라인에서 앞부분은 출근시간(HH:MM:SS), 뒷부분은 퇴근시간이다)

# 특정시간을 입력(예:11:05:20)으로 주었을 때 그 시간에 총 몇 명이 사무실에 있었는지 알려주는 함수를 작성하시오.


# 근무자 수 = 입력한 시간보다 빠르거나 같은 출근 시간 개수 - 입력한 시간보다 빠르거나 같은 퇴근 시간 개수

# 입력: 11:05:20

log_lst ="""
09:12:23 11:14:35
10:34:01 13:23:40
10:34:31 11:20:10
"""

log_lst = log_lst.split()
log_lst = [list(map(int, i.split(":"))) for i in log_lst] # 시, 분, 초 단위로 스플릿
in_lst = log_lst[::2] # 출근 시간들 저장
out_lst = log_lst[1::2] # 퇴근 시간들 저장

input_time = list(map(int, input("시간 입력: ").split(":"))) # 입력 시간 시, 분, 초 단위로 스플릿

def compare_time(time, compare_lst):
    count = 0 # 사람 수
    for j in compare_lst: # 출근 or 퇴근 시간 리스트를 기준으로 입력한 시간과 비교합니다.
        if j[0] == time[0]: # '시'가 같은지
            if j[1] == time[1]: # '분'도 같은지
                if j[2] <= time[2]: # '초'도 같거나 더 빠른지
                    count += 1 # 사람 수 추가
            elif j[1] < time[1]: # '시'가 같을때 '분'이 더 빠른지
                count += 1 # 사람 수 추가
        elif j[0] < time[0]: # '시'가 더 빠른지
            count += 1 # 사람 수 추가
    return count

in_worker = compare_time(input_time, in_lst) # 출근자 수
out_worker = compare_time(input_time, out_lst) # 퇴근자 수
print(in_worker - out_worker) # 남은 근무자 수
```

```python
# 6번 문제 재풀이

# 입력: 11:05:20

logLst = """
09:12:23 11:14:35
10:34:01 13:23:40
10:34:31 11:20:10
"""

logLst = logLst.split()
inTimeLst = logLst[::2]  # 출근 시간들 저장
outTimeLst = logLst[1::2]  # 퇴근 시간들 저장

inputTime = input("시간 입력: ")

def compare_time(time, inTimeLst, outTimeLst):
    # 출근 시간, 퇴근 시간을 한 쌍으로 각각 반복 변수에 할당하여
    # 조건이 참이될 경우 리스트에 1을 담고, 반복문이 끝나면 1을 모두 더한값을 반환합니다.
    return sum([1 for inTime, outTime in zip(inTimeLst, outTimeLst) if inTime <= time < outTime])

print(compare_time(inputTime, inTimeLst, outTimeLst))
```

```python
# 6번 문제 재풀이2

# 입력: 11:05:20

logLst = """
09:12:23 11:14:35
10:34:01 13:23:40
10:34:31 11:20:10
"""

def ret_sec(t): # 초 단위 환산 함수
    hh, mm, ss = t.split(":")
    return int(hh) * 60 * 60 + int(mm) * 60 + int(ss)

logLst = logLst.split()
logLst = [ret_sec(i) for i in logLst]  # 초 단위로 환산하여 리스트 내포

inTimeLst = logLst[::2]  # 출근 시간들 저장
outTimeLst = logLst[1::2]  # 퇴근 시간들 저장

inputTime = ret_sec(input("시간 입력: "))  # 입력 시간 초 단위로 환산

def compare_time(time, inTimeLst, outTimeLst):
    return sum([1 for inTime, outTime in zip(inTimeLst, outTimeLst) if inTime <= time < outTime])

print(compare_time(inputTime, inTimeLst, outTimeLst))
```
