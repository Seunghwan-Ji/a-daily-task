```python
# 1. 사용자 입력과 random함수를 사용하여, 사용자(User)와 컴퓨터가 대결하는 가위 바위 보 게임을 만들어보자.
# 입력: [문자열] "가위", "바위" 혹은 "보" 출력: [문자열] 결과 반환

import random as rd

choiceRps = rd.choice(["가위", "바위", "보"])
inputRps = input("입력: ")

if choiceRps == inputRps:
    print("무승부")
else:
    mc = (choiceRps, inputRps) # (컴퓨터가 낸 것, 내가 낸 것)
    win = [("가위", "바위"), ("바위", "보"), ("보", "가위")] # 이기는 상성
    if mc in win:
        print("승리")
    else:
        print("패배")


# 2. 2^15 = 32768 의 각 자리수를 더하면 3 + 2 + 7 + 6 + 8 = 26 입니다. 2^1000의 각 자리수를 모두 더하면 얼마입니까?
sums = 0
for i in str(2**1000):
    sums += int(i)

print(sums)


# 3. 10진수를 n진수로 변환하는 프로그램을 작성하시오.. (단, n의 범위는 2 <= n <= 16)
# 예)
# 2진수로 변환 : 23310 --> 101101100001110
# 8진수로 변환 : 23310 --> 55416
# 16진수로 변환 : 23310 --> 5b0e

jinsoo16 = {"10": "a", "11": "b", "12": "c", "13": "d", "14": "e", "15": "f"}

def jinsoo(num, n):
    cvs = []
    while True:
        nmg = num % n # 더이상 나눠지지 않을 때까지 나눔
        cvs.append(str(nmg)) # 나눌 때마다 나머지를 리스트에 추가
        mok = int(num / n)
        if mok >= 1: # 몫이 1 이상이면 10진수(num)를 몫으로 대체하고 다시 올라감
            num = mok
            continue
        else: # 나눠지지 않는다면 브레이크
            break
    if n > 10: # 11진수 이상으로(n>=11) 변환하려 할 때 이 과정을 거침
        for i, v in enumerate(cvs):
            if v in jinsoo16: # 요소가 딕셔너리의 키 이름과 같다면
                cvs[i] = jinsoo16[v] # 그 요소를 키 이름으로 바꿈
    return "".join(list(reversed(cvs))) # 결과를 역순으로 읽어서 반환함

jinsoo10 = int(input("10진수 입력: "))
for n in range(2, 17): # 2진수, 3진수, ... , 16진수 출력
    print("{0}진수: {1}".format(n, jinsoo(jinsoo10, n)))


# 4. 문자열 하나를 입력받아 인터넷 주소를 반환하는 make_url 함수를 정의하라.
# make_url("naver")
# www.naver.com

def make_url(url):
    return "www."+url+".com"

make_url("google")


# 5. 문자열을 입력받아 각 문자들로 구성된 리스트로 반환하는 make_list 함수를 정의하라.
# make_list("abcd")
# ['a', 'b', 'c', 'd']

def make_list(strs):
    return [i for i in strs]

make_list("abcd")


# 6. 숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하라.

# pickup_even([3, 4, 5, 6, 7, 8])
# [4, 6, 8]

def pickup_even(numLst):
    return list(filter(lambda e: e % 2 == 0, numLst))

pickup_even([3, 4, 5, 6, 7, 8])


# 7. 콤마가 포함된 문자열 숫자를 입력받아 정수로 변환하는 convert_int 함수를 정의하라.
# convert_int("1,234,567")
# 1234567

def convert_int(strs):
    result = []
    for i in strs:
        try:
            int(i)
        except:
            pass
        else:
            result.append(i)
    return "".join(result)

convert_int("<>1,23!4,@#56가&*?7nㅀ67arㅈ8ㄱ759sㅏ87/^%9힣54$")
```
