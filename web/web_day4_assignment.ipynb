```python
from bs4 import BeautifulSoup
from selenium import webdriver

def url_for_soup(url): # 링크 넣으면 숩 주는 함수
    driver = webdriver.Chrome()
    driver.get(url)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    return soup


# 1) 벅스 뮤직(스크래핑 1위~100위)

# 페이지: "https://music.bugs.co.kr/chart"
# 영역: #CHARTrealtime > table > tbody > tr
    
sp = url_for_soup("https://music.bugs.co.kr/chart")
titles = sp.select("#CHARTrealtime > table > tbody > tr > th") # 영역 > th 태그
artists = sp.select("#CHARTrealtime > table > tbody > tr .artist") # 영역 > 클래스명: artist

for i in range(100):
    print("{0}위".format(i+1))
    print(titles[i].text.strip()) # 곡 제목
    print(artists[i].text.strip()) # 아티스트
    print("="*50)


# 2) 유튜브 랭크(1위~5000위) 추출 - 유튜브 장르, 유튜버, 가입자 수

# 페이지: "https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=" + str(page)

# #list-skin > form:nth-child(4) > table > tbody <- 영역을 가져 왔는데 스크랩이 안되서 아래 영역으로 구현
# 영역: tbody

rank = 1 # 1~5000위

for page in range(1, 51): # 1~50페이지
    sp = url_for_soup("https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page=" + str(page)) # 페이지 패턴
    genre = sp.select("tbody .category") # 영역, 클래스명: category
    name = sp.select("tbody h1 a") # 영역, h1, a
    subscribers = sp.select("tbody .subscriber_cnt") # 영역, 클래스명: subscriber_cnt
    for i in range(100):# 한 페이당 100번
        print("{0}위".format(rank))
        print(genre[i].text.strip()) # 장르
        print(name[i].text.strip()) # 채널 이름
        print(subscribers[i].text.strip()) # 구독자 수
        print("="*50)
        rank += 1


# 3) 보배드림 "국산차" 검색 -> 1페이지~10페이지까지, 페이지당 50개씩, 차량종류, 연식, 주행거리, 가격

# 페이지: "https://www.bobaedream.co.kr/mycar/mycar_list.php?gubun=K&page=" + str(page) + "&order=S11&view_size=50"
# 영역: #listCont > div.wrap-thumb-list > ul

recent = 1 # 최근 1~500번째 매물

for page in range(1, 11): # 1~10페이지
    sp = url_for_soup("https://www.bobaedream.co.kr/mycar/mycar_list.php?gubun=K&page=" + str(page) + "&order=S11&view_size=50") # 페이지 패턴
    name = sp.select("#listCont > div.wrap-thumb-list > ul div.mode-cell.title p.tit") # 영역, mode-cell.title 클래스, tit 클래스
    yay = sp.select("#listCont > div.wrap-thumb-list > ul div.mode-cell.year") # 영역, mode-cell.year 클래스
    mileage = sp.select("#listCont > div.wrap-thumb-list > ul div.mode-cell.km") # 영역, mode-cell.km 클래스
    price = sp.select("#listCont > div.wrap-thumb-list > ul div.mode-cell.price") # 영역, mode-cell.price 클래스
    for i in range(50): # 한 페이지당 50번
        print("%d번째 매물" % (recent))
        print(name[i].text.strip()) # 차 이름
        print(yay[i].text.strip()) # 연식
        print(mileage[i].text.strip()) # 주행거리
        print(price[i].text.strip()) # 가격
        print("="*50)
        recent += 1
```
