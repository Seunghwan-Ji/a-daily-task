```python
# 1. 연속된 세 개의 정수를 더해 12가 되는 경우는 3, 4, 5입니다. 두 정수 num과 total이 주어집니다. 
# 연속된 수 num개를 더한 값이 total이 될 때, 정수 배열을 오름차순으로 담아 return하도록 solution함수를 완성해보세요.

# 제한사항
# 1 ≤ num ≤ 10
# 0 ≤ total ≤ 100
# num개의 연속된 수를 더하여 total이 될 수 없는 테스트 케이스는 없습니다.
# 입출력 예
# num	total	result
# 3	      12	[3, 4, 5]
# 5	      15	[1, 2, 3, 4, 5]
# 4	      14	[2, 3, 4, 5]
# 5	       5	[-1, 0, 1, 2, 3]
# 2	       7	[3,4]
# 1	       5	[5]

# total / num 의 몫은 나누어 떨어지지 않아도 항상 연속되는 정수들 중심에 있는걸 알 수 있습니다.

def solution(num, total):
    standardNum = int(total / num) # 연속되는 정수들 중심에 있는 수 입니다.
    result = [standardNum] # standardNum을 포함한 리스트 생성
    if sum(result) == total: # num이 1일 경우 입니다.
            return sorted(result)
    var = 1 # standardNum 정수에 가감할 변수 입니다.
    while True:
        result.append(standardNum + var) # var값 만큼 더해서 result 리스트에 추가합니다.
        if sum(result) == total:
            return sorted(result) # result 리스트의 요소들 총합이 토탈과 같아지면 정렬된 리스트를 반환하고 함수를 중단합니다.
        else:
            result.append(standardNum - var) # var값 만큼 빼서 result 리스트에 추가합니다.
            if sum(result) == total:
                return sorted(result)
            else:
                var += 1 # var에 1을 더하고 토탈과 값이 같아질 때까지 가감을 다시 반복합니다.

solution(10, 99)


# 2. 네이버 글자수세기 등 특정 글의 글자를 세는 프로그램은 일반적으로 공백을 제외한 글자수만을 세는 기능도 가지고 있다.
# 어떠한 문자열을 입력받았을 때 줄바꿈과 공백을 제외한 글자수만을 리턴하는 코드를 작성하시오.

# 입력 예시
# 공백을 제외한
# 글자수만을 세는 코드 테스트

# 출력 예시
# 18

strs = """
123 4
5         6
                               78
                    9
"""

len("".join(strs.split()))


# 3. 문자와 숫자가섞인 문자열을 입력받을때 구별하여출력해라

# 입력 예시
# "c910m6ia 1ho"

# 출력 예시
# str : cma ho
# int : 91061

import re

strs = "c910m6ia 1ho"

print(*re.findall("[\D]", strs))
print(*re.findall("[\d]", strs))
```
