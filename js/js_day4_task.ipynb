```python
# 1. 직각삼각형의 밑변이 x, 높이가 y일 때 남은 변(대각선)의 길이를 구하는 함수를 만드시오.

def Diagonal_Length(x, y):
    return (x**2 + y**2)**(1/2) # 대각선 길이 제곱의 2분의 1승

inputXY = list(map(float, input("x y: ").split()))
Diagonal_Length(inputXY[0], inputXY[1])


# 2. 기수(Cardinal)를 입력하면 영어 서수(Ordinal)로 출력하는 함수를 작성합니다.

# 1, 21, 31, 41, ... → 1st, 21st, 31st, 41st, ...
# 2, 22, 32, 42, ... → 2nd, 22nd, 32nd, 42nd, ...
# 3, 23, 33, 43, ... → 3rd, 23rd, 33rd, 43rd, ...
# 11, 12, 13, 111, 112, 113, 211, 212, 213, ...  → 11st, 12nd, 13rd, 111st, 112nd, 113rd, 211st, 212nd, 213rd, ...
# 4, 5, 6, 11, 12, 13, 101, 111, 112, ... → 4th, 5th, 6th, 11st, 12nd, 13rd, 101st, 111st, 112nd, ...

# 1로 끝나면 st
# 2로 끝나면 nd
# 3로 끝나면 rd
# 0 or 4이상 th

inputCardinal = input("기수 입력: ").split(", ")
thList = [str(i) for i in range(4, 10)]
thList.append("0")

for i in range(len(inputCardinal)):
    for j in range(len(inputCardinal[i])): # 자릿수 만큼 반복
        try:
            inputCardinal[i][j+1] # ex) n번째 자리 숫자를 보고 있으면 n+1번째 자리에도 숫자가 있는지
        except:
            if inputCardinal[i][j] in thList:
                inputCardinal[i] += "th"
            elif inputCardinal[i][j] == "1":
                inputCardinal[i] += "st"
            elif inputCardinal[i][j] == "2":
                inputCardinal[i] += "nd"
            elif inputCardinal[i][j] == "3":
                inputCardinal[i] += "rd"

print(", ".join(inputCardinal))


# 정규식을 활용한 풀이

import re

inputCardinal = input("기수 입력: ").split(", ")
thList = [str(i) for i in range(4, 10)]
thList.append("0")

for i, v in enumerate(inputCardinal): # 특정 숫자 앞에 0개이상 숫자가 있고 그 특정 숫자로 끝나는지
    if re.match("[0-9]*1$", v):
        inputCardinal[i] = re.sub(v, v+"st", v)
    elif re.match("[0-9]*2$", v):
        inputCardinal[i] = re.sub(v, v+"nd", v)
    elif re.match("[0-9]*3$", v):
        inputCardinal[i] = re.sub(v, v+"rd", v)
    else:
        for j in thList:
            if re.match("[0-9]*"+j+"$", v):
                inputCardinal[i] = re.sub(v, v+"th", v)

print(", ".join(inputCardinal))


# 3.아래 조건에 따라 리스트를 회전하는 프로그램을 작성하시오.

# 조건
# 입력값은 한 행의 문자열로 주어지며, 각 값은 공백으로 구분된다.

# 첫 번째 값은 리스트를 회전하는 양과 방향(음수의 경우 좌측으로, 양수의 경우 우측으로 회전)이다.

# 첫 번째 값을 제외한 나머지 값은 리스트의 각 항목의 값이다.

# 회전된 리스트를 문자열로 출력한다.

# 구현에 이용할 자료구조에 대한 조건이나 제약은 없다.

# 입력되는 리스트의 항목의 개수는 유한한다.

# 입출력예
# 예 1)

# 입력: 1 10 20 30 40 50

# 출력: 50 10 20 30 40

# 예 2)

# 입력: 4 가 나 다 라 마 바 사

# 출력: 라 마 바 사 가 나 다

# 예 3)

# 입력: -2 A B C D E F G

# 출력: C D E F G A B

# 예 4)

# 입력: 0 똘기 떵이 호치 새초미

# 출력: 똘기 떵이 호치 새초미

inputLst = input("정수와 리스트 입력: ").split()
way = int(inputLst[0]) # 방향
inputLst = inputLst[1:] # 리스트

def move(way, lst):
    result = []
    if way != 0:
        for i in range(len(lst)):
            exceedCount = 0 # 인덱스 초과 횟수
            while True:
                try:
                    lst[i-way-exceedCount] # 대체할 값(way가 양의 정수일 경우는 인덱스가 계속 작아질
                except:                    # 것이기 때문에 역순으로 참조되어 오류가 나지 않습니다.)
                    exceedCount += 1 # 초과 횟수 1 증가
                    continue
                else:
                    if not exceedCount: # 초과하지 않았으면 or way가 양의 정수이면
                        result.append(lst[i-way]) # result에 대체할 값 추가
                    else:
                        result.append(lst[exceedCount-1]) # ex) exceedCount가 n이면 n번 초과 이므로
                    break                                 #     대체할 값을 lst의 n-1번째 요소로 선택
    else:
        result = lst
    return result

move(way, inputLst)
```
