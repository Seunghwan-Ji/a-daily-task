```python
# 1. 단일 치환 암호 문제

# 치환된 문자: Q,H,C,B,E,J,K,A,R,W,S,T,U,V,D, ,I,O,P,X,Z,F,G,L,M,N,Y
# 해독할 코드: XHGJQIPWGHZJQPNHD

alphabet_list = [" "] + [chr(i) for i in range(65, 91)] # " ", "A" ~ "Z"
transpose_list = input("치환된 문자: ").split(",")
interpret = input("해독할 코드: ")
conversion_table = {} # 변환표
res = "" # 해독 결과

for i, v in enumerate(transpose_list):
    conversion_table[v] = alphabet_list[i] # 맵핑

for i in interpret:
    res += conversion_table[i]

print(res)


# 2. 2차 배열 정렬 문제

# 1~25 랜덤 배열

import random as rd

board = []
num_record = []

# 1~25까지 중복을 제외한 2차 배열 생성
for _ in range(5):
    row = []
    while len(row) < 5:
        random_num = rd.randint(1, 25)
        if random_num not in num_record:
            row.append(random_num)
            num_record.append(random_num)
    board.append(row)

def res(): # 배열 상태 출력 함수
    for i in board:
        print(*i)

res()

# 1~25 달팽이 배열

# 1~25가 무작위 순서로 담긴 num_record를 정렬
for i in range(len(num_record)-1, 0, -1):
    for j in range(i):
        if num_record[j] > num_record[j+1]:
            temp = num_record[j]
            num_record[j], num_record[j+1] = num_record[j+1], temp
print(num_record)

num = 0 # num_record의 인덱스
row = 0 # 행
col = 0 # 열
record = [] # 숫자 찍은 칸 기록

def move_right(num, row, col, record): # 오른쪽으로 이동
    while num <= 24: # num_record의 인덱스가 25까지 증가했으면 중지
        if col < 5 and (row, col) not in record: # 오른쪽 벽 안 쪽이고, 숫자가 안 찍힌 칸 이면
            board[row][col] = num_record[num] # 숫자 찍기
            record.append((row, col)) # 찍은 칸 기록
            num += 1 # num_record의 인덱스 증가
            col += 1 # 열 증가
        else: # 오른쪽 벽을 뚫었으면
            col -= 1 # 다시 열 위치 원상복귀
            row += 1 # 해당 칸은 숫자를 찍었으니 행 증가
            move_down(num, row, col, record) # 아래로 이동
            break

def move_down(num, row, col, record): # 아래로 이동
    while num <= 24:
        if row < 5 and (row, col) not in record:
            board[row][col] = num_record[num]
            record.append((row, col))
            num += 1
            row += 1
        else: # 바닥을 뚫었으면
            row -= 1 # 행 위치 원상복귀
            col -= 1
            move_left(num, row, col, record) # 왼쪽으로 이동
            break

def move_left(num, row, col, record): # 왼쪽으로 이동
    while num <= 24:
        if col >= 0 and (row, col) not in record:
            board[row][col] = num_record[num]
            record.append((row, col))
            num += 1
            col -= 1
        else: # 왼쪽 벽을 뚫었으면
            col += 1 # 열 위치 원상복귀
            row -= 1
            move_up(num, row, col, record) # 위로 이동
            break

def move_up(num, row, col, record): # 위로 이동
    while num <= 24:
        if row >= 0 and (row, col) not in record:
            board[row][col] = num_record[num]
            record.append((row, col))
            num += 1
            row -= 1
        else: # 천장을 뚫었으면
            row += 1 # 행 위치 원상복귀
            col += 1
            move_right(num, row, col, record) # 오른쪽으로 이동
            break

move_right(num, row, col, record)
res()
```
